```python
import pandas as pd

data = pd.read_excel("ESP.xlsx")
display(data.head(10))
```

```python
# to get data summary

gp = data.groupby("Department").agg({"Department" : "count"})
display(gp)
```

```python
gp1 = data.groupby(["City", "Department"]).agg({"Bonus %" : "max"})
display(gp1)
```

```python
gp1 = data.groupby(["City", "Department"]).agg({"Bonus %" : "min"})
display(gp1)
```

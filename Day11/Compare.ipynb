```python
import pandas as pd

data1 = {"Names" : ["Ram", "Shyam", "Rahul"],
         "Age" : [34, 56, 43]}

df1 = pd.DataFrame(data1)
display(df1)

df2 = df1.copy()
display(df2)

df2.loc[0, "Names"] = "Shaur"
df2.loc[1, "Names"] = "Rudra"
df2.loc[2, "Names"] = "Rahul"
df2.loc[0, "Age"] = 20
df2.loc[1, "Age"] = 21
df2.loc[2, "Age"] = 22

display(df2)

# we can visually compare in small data but when we have big data it becomes tough to look through everything so thats why we use compare function
```

```python
display(df1.compare(df2, align_axis=0))

display(df1.compare(df2))

display(df1.compare(df2, keep_equal=True))
display(df1.compare(df2, keep_equal=False))
```

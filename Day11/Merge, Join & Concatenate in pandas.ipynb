```python
import pandas as pd

data1 = {"Emp ID" : ["E01", "E02", "E03", "E04", "E05", "E06"],
         "Names" : ["Ram", "Shyam", "Rahul", "Vishal", "Ravi", "John"],
         "Age" : [34, 56, 43, 23, 45, 22]}
 
data2 = {"Emp ID" : ["E01", "E07", "E03", "E04", "E08", "E06"],
         "Salary" : [45000, 40000, 60000, 75000, 50000, 100000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
display(pd.merge(left = df1, right = df2, on = "Emp ID", how = "left"))
```

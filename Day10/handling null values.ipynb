```python
import pandas as pd

data = pd.read_csv("null.csv")
display(data)
```

```python
data.isnull().sum()
```

```python
display(data.dropna())
```

```python
import numpy as np
display(data.replace(np.nan, "hi"))     # replaces null values with hi, but we don't want hi in attendance or anything else
```

```python
data["City"] = data["City"].replace(np.nan, "Chennai")
display(data)
```

```python
display(data.fillna(method = "bfill"))      # backward fill : the value is backwarded to the null value
```

```python
display(data.fillna(method = "ffill"))      #forward fill : the value gets forwarded to the null value
```

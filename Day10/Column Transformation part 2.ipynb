```python
import pandas as pd

data = pd.read_csv("small_sample.csv")
display(data)

data["Marks Deducted"] = (100 - data["Marks"])
display(data)
```

```python
data = {"Months" : ["January", "February", "March", "April"]}

a = pd.DataFrame(data)
print(a)

def extract(value):
    return value[0:3]

a["Short Months"] = a["Months"].map(extract)        #method to create a new column using an existing column
print(a)
```

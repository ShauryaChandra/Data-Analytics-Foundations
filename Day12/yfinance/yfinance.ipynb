```python
import yfinance as yf
import pandas as pd

df = yf.download("AAPL", start = "2025-01-01", end = "2025-12-31")
display(df.head())
display(df.shape)
```

```python
import matplotlib.pyplot as plt

plt.plot(df.index, df["Close"])
plt.title("AAPL closing price of 2025")
plt.xlabel("Date")
plt.ylabel("price (USD)")
plt.savefig("AAPL_Price.png")
plt.show()
```

```python
# calculate daily returns

print(df.columns)
print(df.head())
df = yf.download("AAPL", start="2023-01-01", end="2024-01-01")
df.columns = df.columns.get_level_values(0)


df["Return"] = (df["Close"] - df["Close"].shift(1)) / df["Close"].shift(1) * 100
display(df["Return"].head(10))
```

```python
#plotting returns as histogram

plt.hist(df["Return"].dropna(), bins = 50, edgecolor = "black")
plt.title("AAPL daily returns")
plt.xlabel("Daily Return %")
plt.ylabel("Frequency")
plt.savefig("aapl_returns.png")
plt.show()
```

```python
tickers = ["AAPL", "TSLA", "MSFT"]

prices = yf.download(tickers, start = "2025-01-01", end = "2025-12-31")

for ticker in tickers:
    plt.plot(prices.index, prices["Close"][ticker], label = ticker)

plt.title("Stock Prices 2025")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.savefig("three_stocks.png")
plt.show()
```

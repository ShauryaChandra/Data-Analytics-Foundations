```python
import matplotlib.pyplot as plt

x = ["Day1", "Day2", "Day3", "Day4", "Day5"]

y = [300, 340, 230, 270, 467]
y1 = [350, 370, 300, 310, 534]
plt.plot(x,y, marker = "o", ls = ":", color = "Red", label = "Week 1")
plt.plot(x,y1, marker = "*", ls = "--", color = "Blue", label = "Week2")
plt.legend()
plt.xlabel("Days")
plt.ylabel("Customers")
plt.title("Customers per day in a restaurant")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("expense1.csv")
display(data)

df = data.groupby("Month")["Amount"].sum()
#df = data.groupby("Month")["Amount"].sum()
display(df)

plt.plot(df.index, df.values)

plt.title("Total Expense by Month")
plt.xlabel("Month")
plt.ylabel("Amount")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Part 1", "Part 2", "Part 3", "Part 4", "Part 5"]
y = [98, 67, 88, 95, 88]

colors = ["red", "green", "blue", "yellow", "purple"]
plt.bar(x, y, color = colors, edgecolor = "orange")
plt.xlabel("Parts of harry potter", fontsize = 15)
plt.ylabel("Popularity", fontsize = 15)
plt.title("Popularity of different parts of harry potter", fontsize = 15)
plt.show()
```

```python
import numpy as np
import statistics as stats

baked_food = [200,300,150,130,200,280,170,188]      # Carbohydrate quantity of baked food

a = np.array(baked_food)
print(np.mean(baked_food))      #sum of all values/number of values
print(np.median(baked_food))    #central value after sorting
# there is no function for mode so we use statistics library
print(stats.mode(baked_food))   #most frequently occurring value in the data
print(np.std(baked_food))       #high std means that the data is zyaada spread and so the fluctuations are high in the product. (mostly used in quality control)
print(np.var(baked_food))       #square of std
```

```python
# -1 represents inversely proportional relationship
# 1 represents proportional relationships
# 0 means no relationship
tobacco_consumption = [20,50,10,30,50,40]
deaths = [100, 120, 70, 100, 120, 112]

print(np.corrcoef(tobacco_consumption, deaths))     # shows if tobacco consumption rises then deaths also rises
```

```python
price = [300,100,350,150,200]
sales = [10,20,7,17,3]

print(np.corrcoef(price, sales))        # shows if price increases, sales decreases
```

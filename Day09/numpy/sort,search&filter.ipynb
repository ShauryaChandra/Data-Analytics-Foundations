```python
import numpy as np
arr = np.array([13,41,2,4,78,23])
print(np.sort(arr))
```

```python
s = np.where(arr == 4)
print(s)

S = np.where(arr % 2 == 0)
print(S)
```

```python
import numpy as np
ar = np.array([2,3,4,5,6])
ss = np.searchsorted(ar, 4)  #Only use on arrays which are already sorted
print(ss)
```

```python
#Filter
import numpy as np
arr = np.array([20,30,40,50])
fa = [True, False, True, False]
new = arr[fa]
print(new)
```
